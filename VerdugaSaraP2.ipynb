```python
import pandas as pd

# 1) Cargar dataset
df = pd.read_csv("heart.csv")

# 2) Vista rápida
print("✅ Filas, Columnas:", df.shape)
print("\n✅ Primeras 5 filas:")
print(df.head())

# 3) Tipos y nulos
print("\n✅ Info del dataset:")
print(df.info())

print("\n✅ Nulos por columna:")
print(df.isnull().sum())

# 4) Estadísticas básicas (solo numéricas)
print("\n✅ Describe:")
print(df.describe(include="number"))
```
